```python
from algorithm import BranchAndBound

problem = {
    "objective": {
        "r1": 6, # menit
        "r2": 8, # menit
        "r3": 10 # menit
    },
    "constraints": [
        # pilih minimal satu rute
        ({"r1": 1, "r2": 1, "r3": 1}, ">=", 1),

        # batas keparahan banjir
        ({"r1": 5, "r2": 2, "r3": 3}, "<=", 3),

        # binary constraints agar hasil 0 atau 1 saja
        ({"r1": 1}, "<=", 1),
        ({"r2": 1}, "<=", 1),
        ({"r3": 1}, "<=", 1),
    ],
    "variables": ["r1", "r2", "r3"]
}


bnb = BranchAndBound(problem)
result = bnb.solve()

print("RESULT:")
print(result)
```
